# Exploratory Data Analysis

# Gender

```python
#plotting bar chart for gender and its count

gen=sns.countplot(x='Gender',data=df)

#adding count to that
for bars in gen.containers:
    gen.bar_label(bars)
```

```python
df.groupby('Gender')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart to show total amounts sum based on gender(gender vs amount)


sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
#From above graphs we can see that most of the buyers are females and even the count of females are greater than men
```


# Age

```python
ax=sns.countplot(x='Age Group',data=df,hue ='Gender')

#adding count to that
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sadddleofgen = df.groupby('Gender')['Amount'].sum().reset_index(name='Total Amount').sort_values(by='Total Amount', ascending=False)
```

```python
#grouping the total amount according to the age group(age group vs amount)

agesales = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(15,6)})

sns.barplot(x='Age Group', y='Amount', data=agesales)
```

```python
#from the above we can say that the maximum age group is between 26-35 and they are females by a large majority
```

```python
df.columns
```

# State

```python
st=sns.countplot(data=df,x='State')
sns.set(rc={'figure.figsize':(15,4)})

for bars in st.containers :
    st.bar_label(bars)
    
```

```python
#now we will evaluate the sales in each state by the amount

statesales = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(x='State', y='Amount', data=statesales)
```

```python
#total number of orders from each state (state vs order)
stateorders = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(x='State', y='Orders', data=stateorders)
```

```python
#hence we can see that most of the states with high orders have high amount spendings, except in the case of Kerala and Haryana.
```

```python
#UP ,Maharashtra ,Karnatak and Delhi have the most sales orders and purchases.
```

```python
df.columns
```

# Marital Status 

```python
mt= sns.countplot(x='Marital_Status',data =df,hue ='Gender')

for bars in mt.containers:
    mt.bar_label(bars)
```

```python
#most of the orders are from married woman
```

```python
mar_sales= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
sns.barplot(x='Marital_Status',y='Amount',data=mar_sales,hue='Gender')
```

```python
#Most of the buyers are woman and they are married
```

# Occupation

```python
occ=sns.countplot(data=df,x='Occupation')
for bars in occ.containers:
    occ.bar_label(bars)
```

```python
#finding buyers from the occupation (occupation vs amount)

occ_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=occ_sales,x='Occupation',y='Amount')
```

```python
#Hence most of the buyers are from IT sector,Healthcare and  aviation department
```

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
#hence the maximum products sold products are from food ,clothing, and electronics
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# the top 10 most sold products

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
